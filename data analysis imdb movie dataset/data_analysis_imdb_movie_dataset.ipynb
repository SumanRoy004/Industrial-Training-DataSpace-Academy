# 20. Classify Movies Based on Ratings [Good,Better and Best]

```python
s=pd.Series(data['Rating'])
k2=s.mode().loc[0]
k1=(s.max()+s.min())/2
(k1,k2)
```

```python
# Rating<=k1--->Good
# k1<Rating<k2---->Better
# k2<=Rating----->Best
data['Class']=['Good' if x<=k1 else 'Better' if k1<x and x<k2 else 'Best' for x in data['Rating']]       # New column addition
```

```python
_=sns.displot(x='Rating',data=data,kind='kde',hue='Class')
```
